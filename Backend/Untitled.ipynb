```python
import PySimpleGUI as sg
import cv2
import numpy as np
import sys
from sys import exit as exit
from datetime import datetime
import requests
import pandas as pd
from tkinter import messagebox
import csv
```

```python
def leer_placa(img):
    regions = ['gb','it']
    with open(img, 'rb') as fp:
        
        response = requests.post(
            'https://api.platerecognizer.com/v1/plate-reader/',
            data=dict(regions=regions),  # Optional
            files=dict(upload=fp),
            headers={'Authorization': 'Token 5692caed871a24a4b3e2fd9594eb0f73146c4144'})
    return response.json()

            
max_num_plate = 10
sg.ChangeLookAndFeel('Black')
layout = [[sg.Text('PLATE by Vania Vera', size=(40, 1), justification='center', font='Helvetica 20')],
          [sg.Image(filename='', key='image')],
          [sg.Button('Read plate', size=(10, 1), font='Helvetica 14'),
           sg.Button('Map', size=(10, 1), font='Any 14'),
           sg.Button('Exit', size=(10, 1), font='Helvetica 14')]]
window = sg.Window('Demo Application - Read Plate v1.1', layout,
                   location = (800,400))

cap = cv2.VideoCapture(0)
recording = True
while True:
    event, values = window.Read(timeout=20)
    if event == 'Exit' or event is None:
        break
    elif event == 'Read plate':
        break 
    elif event == 'Map':
        break
    if recording:
        ret, frame = cap.read()
        imgbytes = cv2.imencode('.png', frame)[1].tobytes()
        window.FindElement('image').Update(data=imgbytes)
window.close()

exit()
        
```
